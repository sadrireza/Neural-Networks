Step 2: Load Pre-trained Transformer Model

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the pre-trained T5 model and tokenizer
tokenizer = T5Tokenizer.from_pretrained('t5-base')
model = T5ForConditionalGeneration.from_pretrained('t5-base')
```

Step 3: Load Text File

```python
# Upload your text file to Google Colab
from google.colab import files
uploaded = files.upload()

# Get the file name
filename = list(uploaded.keys())[0]

# Read the text file
with open(filename, 'r') as f:
    text = f.read()
```

Step 4: Preprocess Text

```python
# Tokenize the text
input_ids = tokenizer.encode_plus(
    text,
    max_length=512,
    padding='max_length',
    return_tensors='pt'
)
```

Step 5: Syntax Correction

```python
# Generate the corrected text
output = model.generate(
    input_ids['input_ids'],
    max_length=512,
    num_beams=5,
    early_stopping=True
)

# Decode the corrected text
corrected_text = tokenizer.decode(output[0], skip_special_tokens=True)
```

Step 6: Print Corrected Text



```python
print('Original Text:')
print(text)

print('\nCorrected Text:')
print(corrected_text)
```
